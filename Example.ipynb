Barmat: Julia example
------------------------
Recreate a plot from Gao's 2008 Caltech Thesis (figure 2.5)

```julia
#Load up the Julia code
include("./src/Barmat.jl")

#This is a Matplotlib wrapper for Julia
using PyPlot
```

```julia
#Material parameters for Al from Pöpel, 1989 (his reference 19), except 2*bcs = 3.4
bcs = 3.4/2.0
tc = 1.178
london0 = 15.4e-9
vf=1.34e6
mfp = 10e-6

#These can be calculated using the built-in convenience functions
delta0 = Barmat.get_delta0(tc, bcs)

#Pöpel uses a table value that is slightly different, so results may not be exact
ksi0 = Barmat.get_ksi0(vf, delta0)

#Vector of interesting temperatures
trs = [0.5, 0.7, 0.8, 0.9]

#Since kB*Tc = delta0/bcs and fr = h*f/delta0, then h*f/kB*Tc = fr/bcs
frs = linspace(0, 8, 50)/bcs; #<-- semicolon suppresses this guy from getting written to the screen
```

```julia
#Empty dict to hold results as function of temps
zsd = Dict()

for tr in trs
    zsd[tr] = Barmat.get_Zvec(frs, tc, vf, london0, "frequency", tr=tr, mfp=mfp, 
                              bcs=bcs, output_depths = true, boundary = "d")
end
```

```julia
#The notation here looks crazy if you come from Python. But it works.

fig, axes = subplots(1, 2)
fig[:set_size_inches](12, 6)
ax1 = axes[1]
ax2 = axes[2]


for tr in trs
    line, = ax1[:plot](frs*bcs, 1e9*real(zsd[tr]))
    ax2[:plot](frs*bcs, 1e9*imag(zsd[tr]), color=line[:get_color](), label="T/Tc = $tr")
end


ax1[:set_xlabel](L"frequency ($hf/k_\mathrm{B}T_\mathrm{c}$)")
ax2[:set_xlabel](L"frequency ($hf/k_\mathrm{B}T_\mathrm{c}$)")
ax1[:set_ylabel](L"Skin depth $\delta_\mathrm{s}$ (nm)")
ax2[:set_ylabel](L"Penetration depth $\lambda_\mathrm{eff}$ (nm)")
ax2[:legend](loc="best")
fig[:tight_layout]()
```
